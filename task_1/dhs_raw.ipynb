```python
#Import libraries.
import json
import urllib3 as urllb
import pandas as pd

http = urllb.PoolManager()
```

## Countries

```python
#Query the API for a list of countries.
countries_url = r'https://api.dhsprogram.com/rest/dhs/countries'

#Obtain and Parse the list into a Python Object.
req = http.request('GET', countries_url)
resp = json.loads(req.data.decode('utf-8')) # json.loads(req.data)

# #Get the list of countries, put in dataframe.
countries =  pd.DataFrame(resp['Data'])
countries
```

## Indicators

```python
indicators =pd.read_csv('dhs_indicators.csv')
indicators.head()
```

```python
# http = urllb.PoolManager()

# url https://api.dhsprogram.com/rest/dhs/indicators
indicator_url = r'https://api.dhsprogram.com/rest/dhs/indicators'

#Obtain and Parse the list into a Python Object.
req = http.request('GET', indicator_url)
resp = json.loads(req.data.decode('utf-8')) 

# get the list of indicators, put in dataframe.
indicators =  pd.DataFrame(resp['Data'])
indicators
```

```python
indicators.to_csv('dhs_indicators.csv', index=False)
```

```python
# filter indicators with "malaria" in the definition
malaria_indicators = indicators[indicators['Definition'].str.contains('malaria', case=False)].reset_index(drop=True)
malaria_indicators
```

```python
# filter indicators with "fever" in the definition
fever_indicators = indicators[indicators['Definition'].str.contains('fever', case=False)].reset_index(drop=True)
fever_indicators
```

```python
# aggregate the IndicatorId in malaria_indicators and fever_indicators, drop duplicates
indicator_ids = pd.concat([malaria_indicators['IndicatorId'], fever_indicators['IndicatorId']]).drop_duplicates().reset_index(drop=True)
indicator_ids
```

```python
# http = urllb.PoolManager()

# https://api.dhsprogram.com/rest/dhs/data/FE_FRTR_W_TFR,subnational
indicator_ids_url = r'https://api.dhsprogram.com/rest/dhs/data/subnational,' + ','.join(indicator_ids.astype(str))
# indicator_ids_url = r'https://api.dhsprogram.com/rest/dhs/data/subnational,' + ','.join(indicator_ids.iloc[:2].astype(str))
indicator_ids_url

#Obtain and Parse the list into a Python Object.
req = http.request('GET', indicator_ids_url)
resp = json.loads(req.data.decode('utf-8'))

# get the list of responses, put in dataframe.
indicator_subntl_data =  pd.DataFrame(resp['Data'])
indicator_subntl_data
```

```python
# iterate through the indicator_ids and get the data for each indicator
indicator_subntl_data = pd.DataFrame()
for indicator_id in indicator_ids:
    indicator_ids_url = r'https://api.dhsprogram.com/rest/dhs/data/subnational,' + str(indicator_id)
    req = http.request('GET', indicator_ids_url)
    resp = json.loads(req.data.decode('utf-8'))
    indicator_subntl_data = pd.concat([indicator_subntl_data, pd.DataFrame(resp['Data'])], axis=0)  
indicator_subntl_data
```

```python
# export the data to csv
indicator_subntl_data.to_csv('dhs_subntl_data_raw.csv', index=False)
```
