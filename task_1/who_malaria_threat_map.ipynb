```python
import pandas as pd
```

```python
## this helps preparing ISO2 and ISO3 codes for countries

# read country_iso_code csv from data folder, keep 'country', 'iso2' and 'iso3' columns, turn off na_values (since "NA" is a valid value for Namibia)
country_iso_code = pd.read_csv('../data/country_iso_code.csv', usecols=['country', 'iso2', 'iso3'], na_values=None)
```

```python
# read the WHO Malaria Threat Map excel file from data folder
# NA values are either empty cells or cells with "NR" string
mtm = pd.read_excel('../data/MTM_HRP2_map_13_01_23.xlsx', sheet_name='Deletions_13.1.23', na_values=['', 'NR'])
# mtm.head()
```

```python
# some of the samples were taken from travelers, so we need to separate them from the rest
# we will use the WHO_SAMPLE_ORIGIN_TEXT column to do that
# whichever row where WHO_SAMPLE_ORIGIN_TEXT is not empty, we will put it in a separate dataframe
# create a new dataframe for travelers
travelers = mtm[mtm.WHO_SAMPLE_ORIGIN_TEXT.notnull()]
travelers
```

```python
# export travelers to csv file
# travelers.to_csv('../data/mtm_2023_01_13_sheet-travelers.csv', index=False, encoding='utf-8')
```

```python
# update mtm dataframe to exclude travelers
mtm = mtm[mtm.WHO_SAMPLE_ORIGIN_TEXT.isnull()]

# export mtm to csv file
# mtm.to_csv('../data/mtm_2023_01_13_sheet-indigenous_samples.csv', index=False, encoding='utf-8')

mtm.head(n=10)
```

```python
# columns to keep in mtm:
# UNIQUE_ID, ISO_ctry, WHO_ADMIN1, WHO_LATITUDE, WHO_LONGITUDE, WHO_YEAR_START, 
# WHO_PATIENTS_SCREENED, ## this is the total number of patients screened
# WHO_SURVEY_SIZE_CASES, ## this is the total number of confirmed Pf cases
# WHO_PF_POS_SAMPLES, ## this is the total number of positive samples in the survey
# WHO_PF_POS_RATE, ## this is percentage of positive cases
# WHO_PATIENT_TYPE, ## this has one of these values: Asymptomatic, Symptomatic, Mixed symptomatic and asymptomatic
# WHO_PF_POS_RATE_METHOD, ## this has one of these values: HRP2 based RDT, nonHRP2 based RDT, microscopy, PCR, other
# POSITIVITY_METHOD_2, ## this has one of these values: HRP2 based RDT, nonHRP2 based RDT, microscopy, PCR, other
# POSITIVITY_METHOD_3, ## this has one of these values: HRP2 based RDT, nonHRP2 based RDT, microscopy, PCR, other
# WHO_RDT_TYPE, ## name of RDT brand used for testing
# RDT_TYPE_2, ## name of RDT brand used for testing 
# WHO_HRP2_DELETION_TESTED_N, 
# WHO_HRP2_DELETION_POS_N, 
# WHO_HRP2_DELETION_PERCENT, 
# WHO_HRP2_ONLY, ## 1 = HRP2 analysed, HRP3 not analysed; 0 = HRP2  analysed and HRP3 analysed; 2 = HRP2 analysed and HRP3 deemed not appropriate to analyse
# WHO_HRP3_DELETION_TESTED_N, 
# WHO_HRP3_DELETION_POS_N, 
# WHO_HRP3_DELETION_PERCENT, 
# WHO_HPR2_HRP3_DELETIONS_TESTED_N, 
# WHO_HPR2_HRP3_DELETIONS_POS_N, 
# WHO_HPR2_HRP3_DELETIONS_PERCENT, 
# WHO_STANDARD_METHOD, ## yes if HRP2 analysis was done per WHO recommendations
# WHO_SERO_RESULTS, ## this has one of these values: HRP2 present, HRP2 absent, Confirm PCR results, Not performed
# DELETION_STATUS ## this can be either confirmed or suspected

# keep only the columns we need
# mtm = mtm[['UNIQUE_ID', 'ISO_ctry', 'WHO_ADMIN1', 'WHO_LATITUDE', 'WHO_LONGITUDE', 'WHO_YEAR_START', 'WHO_PATIENTS_SCREENED', 'WHO_SURVEY_SIZE_CASES', 'WHO_PF_POS_SAMPLES', 'WHO_PF_POS_RATE', 'WHO_PATIENT_TYPE', 'WHO_PF_POS_RATE_METHOD', 'POSITIVITY_METHOD_2', 'POSITIVITY_METHOD_3', 'WHO_RDT_TYPE', 'RDT_TYPE_2', 'WHO_HRP2_DELETION_TESTED_N', 'WHO_HRP2_DELETION_POS_N', 'WHO_HRP2_DELETION_PERCENT', 'WHO_HRP2_ONLY', 'WHO_HRP3_DELETION_TESTED_N', 'WHO_HRP3_DELETION_POS_N', 'WHO_HRP3_DELETION_PERCENT', 'WHO_HPR2_HRP3_DELETIONS_TESTED_N', 'WHO_HPR2_HRP3_DELETIONS_POS_N', 'WHO_HPR2_HRP3_DELETIONS_PERCENT', 'WHO_STANDARD_METHOD', 'WHO_SERO_RESULTS', 'DELETION_STATUS']]
# mtm.head(n=10)
```

```python
# the column WHO_YEAR_START in mtm is a string, 
# if the string has more than 4 characters, keep only the last 4 characters,
# else keep the whole string
mtm['WHO_YEAR_START'] = mtm['WHO_YEAR_START'].apply(lambda x: x[-4:] if len(str(x)) > 4 else x)
# convert WHO_YEAR_START to integer type
mtm['WHO_YEAR_START'] = pd.to_numeric(mtm['WHO_YEAR_START'], errors='coerce')
# mtm.head(n=10)
```

```python

# sort mtm by ISO_ctry then WHO_YEAR_START
mtm = mtm.sort_values(by=['ISO_ctry', 'WHO_ADMIN1', 'WHO_YEAR_START'])
# mtm.head(n=10)
```

```python
# export mtm to csv file, NA values are empty cells
# mtm.to_csv('../data/mtm_2023_01_13_sheet-indigenous_samples.csv', index=False, na_rep='')

# read mtm_2023_01_13_sheet.csv from data folder
# mtm = pd.read_csv('../data/mtm_2023_01_13_sheet.csv')
# mtm.head(n=10)
```

## Symptomatic cases

```python
# subset mtm to include rows where WHO_PATIENT_TYPE is Symptomatic
mtm_symp = mtm[mtm.WHO_PATIENT_TYPE == 'Symptomatic']
```

## HRP2, HRP3 samples

```python
# prepare dataframe to output to csv file
df = pd.DataFrame()
df['iso2'] = mtm_symp['ISO_ctry']
df['admin1'] = mtm_symp['WHO_ADMIN1']
df['latitude'] = mtm_symp['WHO_LATITUDE']
df['longitude'] = mtm_symp['WHO_LONGITUDE']
df['year_start'] = mtm_symp['WHO_YEAR_START']

df['hrp2_screened'] = mtm_symp['WHO_HRP2_DELETION_TESTED_N']
df['hrp2_pos'] = mtm_symp['WHO_HRP2_DELETION_POS_N']
df['hrp2_pos_frac'] = mtm_symp['WHO_HRP2_DELETION_POS_N'] / mtm_symp['WHO_HRP2_DELETION_TESTED_N'] # mtm_symp['WHO_HRP2_DELETION_PERCENT']

df['hrp3_screened'] = mtm_symp['WHO_HRP3_DELETION_TESTED_N']
df['hrp3_pos'] = mtm_symp['WHO_HRP3_DELETION_POS_N']
df['hrp3_pos_frac'] = mtm_symp['WHO_HRP3_DELETION_POS_N'] / mtm_symp['WHO_HRP3_DELETION_TESTED_N'] # mtm_symp['WHO_HRP3_DELETION_PERCENT']

df['hrp2_hrp3_screened'] = mtm_symp['WHO_HPR2_HRP3_DELETIONS_TESTED_N']
df['hrp2_hrp3_pos'] = mtm_symp['WHO_HPR2_HRP3_DELETIONS_POS_N']
df['hrp2_hrp3_pos_frac'] = mtm_symp['WHO_HPR2_HRP3_DELETIONS_POS_N'] / mtm_symp['WHO_HPR2_HRP3_DELETIONS_TESTED_N'] # mtm_symp['WHO_HRP2_HRP3_DELETION_PERCENT']

df['src'] = 'WHO Malaria Threat Map'
df['src_pmc'] = mtm_symp['WHO_PMC']
df['src_url'] = mtm_symp['WHO_CITATION_URL']
# look up iso3 code from iso2 code, using country_iso_code dataframe
df['iso3'] = df['iso2'].apply(lambda x: country_iso_code[country_iso_code.iso2 == x].iso3.values[0])
# look up country name from iso2 code, using country_iso_code dataframe
df['country'] = df['iso2'].apply(lambda x: country_iso_code[country_iso_code.iso2 == x].country.values[0])

df.head(n=10)
```

```python
# df admin0
df_admin0 = df.groupby(['iso2', 'year_start']).agg({'hrp2_screened': 'sum', 'hrp2_pos': 'sum', 'hrp3_screened': 'sum', 'hrp3_pos': 'sum', 'hrp2_hrp3_screened': 'sum', 'hrp2_hrp3_pos': 'sum'}).reset_index()

df_admin0['hrp2_pos_frac'] = df_admin0['hrp2_pos'] / df_admin0['hrp2_screened']
df_admin0['hrp3_pos_frac'] = df_admin0['hrp3_pos'] / df_admin0['hrp3_screened']
df_admin0['hrp2_hrp3_pos_frac'] = df_admin0['hrp2_hrp3_pos'] / df_admin0['hrp2_hrp3_screened']

df_admin0['iso3'] = df_admin0['iso2'].apply(lambda x: country_iso_code[country_iso_code.iso2 == x].iso3.values[0])
df_admin0['country'] = df_admin0['iso2'].apply(lambda x: country_iso_code[country_iso_code.iso2 == x].country.values[0])
df_admin0['admin1'] = 'none'
df_admin0['latitude'] = ''
df_admin0['longitude'] = ''
df_admin0['src'] = 'WHO Malaria Threat Map'
df_admin0['src_pmc'] = ''
df_admin0['src_url'] = ''
df_admin0.head(n=10)
```

```python
# concatenate df_admin0 to df
df = pd.concat([df, df_admin0], ignore_index=True)

# sort df by iso2, year_start, admin1
df = df.sort_values(by=['iso2', 'year_start', 'admin1'])

df.head(n=10)
```

```python
# rearrange columns in df then export to csv file
df = df[['iso2', 'iso3', 'country', 'admin1', 'latitude', 'longitude', 'year_start', 'hrp2_screened', 'hrp2_pos', 'hrp2_pos_frac', 'hrp3_screened', 'hrp3_pos', 'hrp3_pos_frac', 'hrp2_hrp3_screened', 'hrp2_hrp3_pos', 'hrp2_hrp3_pos_frac', 'src', 'src_pmc', 'src_url']]
# export df to csv file
df.to_csv('mtm_2023_01_13_sheet-hrp2_3_indigenous_symptomatic_samples.csv', index=False, na_rep='')
```
